```python
missing_values = ['.','%','a','?','NaN']
new_data = pd.read_excel('data_preprocessing.xlsx', na_values=missing_values)
data1 = new_data.copy()
```

```python
data1
```

```python
m_Age = (data1['Age'].median())
m_Age
```

```python
print(data1['Age'].unique())
data1['Age'].fillna(value = data1['Age'].median(), inplace = True)
data1['Age'].replace(to_replace=0.0, value =  data1['Age'].median(), inplace = True )
```

```python
print(data1['Salary'].unique())
data1['Salary'].fillna(value = data1['Salary'].median(), inplace = True)
```

```python
print(data1['Experience'].unique())
data1['Experience'].fillna(value = data1['Experience'].median(), inplace = True)
```

```python
print(data1['Dept'].unique())
```

```python
data1['Dept'].value_counts()
```

```python
print(data1['Education'].unique())
condition = data1['Salary']<30000.0
data1['Education'] = data1['Education'].fillna(condition.map({True:'Under_Graduate', False:'Master'}))
```

```python
print(data1['Education'].unique())
```

```python
print(data1['Projects'].unique())
data1['Projects'].fillna(value = data1['Projects'].median(), inplace = True)
```
