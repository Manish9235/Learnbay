```python
import pandas as pd
```

```python
data = {'names': ['Alice', 'John', 'Mark', 'Carol'],
       'ids': [101, 102, 103, 104],
       'percents': [78.5, 87.5, 68.0, 91.5]}
```

```python
df = pd.DataFrame(data, columns=['ids', 'names', 'percents'])
```

```python
df
```

```python
# Create a new column - 'grades'
# Values:
# percent > 90 => 'A'a
# 90 > percent > 80 => 'B'
# 80 > percent > 70 => 'C'
# 70 > percent > 60 => 'D'
# percent < 60 => 'F'
```

```python
def get_grade(marks):
    if marks >= 90:
        return 'A'
    elif 90 > marks >= 80:
        return 'B'
    elif 80 > marks >= 70:
        return 'C'
    elif 70 > marks >= 60:
        return 'D'
    else:
        return 'F'
```

```python
df['grades'] = df['percents'].apply(get_grade)
```

```python
df
```

```python
def get_grade(marks, a, b):
    if marks < 60:
        return marks + a + b
    else:
        return marks + 1
```

```python
df['col1'] = df['percents'].apply(get_grade, args=(2, 5))
```
